```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
z = np.array([-5,-4,-3,-2,-1,1,2,3,4,5])
z = np.sort(z)
```

# Sigmoid Function

    

```python
a = 1/(1+np.exp(-z))
plt.plot(a)

#Derivative of sigmoid function (say a) a' = a(1-a)
aprime = a*(1-a)
plt.plot(aprime)
```

# Tanh Function

```python
a = (np.exp(z)-np.exp(-z))/(np.exp(z)+np.exp(-z))
plt.plot(a)

#Derivative of sigmoid function (say a) a' = (1-a^2)
aprime = (1-a*a)
plt.plot(aprime)
```

# Relu Function

```python
a = np.maximum(0,z)
plt.plot(a)

#Derivative of sigmoid function (say a) a' = {0,if a<0 and a if a>0}
aprime = np.where (a>0, a,0)
plt.plot(aprime)
```

# Leaky Relu

    - Under normal circumstances if z<0 we would multiply it with 0.001. However when plotting it using matplotlib the graph was not coming out well and i thought it's a bug !! Hence for representation purposes only i am multiplying it using 0.1

```python
a = np.where(z>0,z,z*0.1)
plt.plot(a)
```
